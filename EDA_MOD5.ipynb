```python
trace_show = go.Scatter(
    x=month_show.Date,
    y=month_show.Show_Impression,
    name = "Show-Impression",
    line = dict(color = '#17BECF'),
    opacity = 0.8)
trace_comm = go.Scatter(
    x=month_commercial.Date,
    y=month_commercial.Commercial_Impression,
    name = "Comercial-Impression",
    line = dict(color = '#7F7F7F'),
    opacity = 0.8)


data = [trace_show, trace_comm]

layout = dict(
    title='Time Series with Rangeslider',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),
        type='date'
    )
)

fig = dict(data=data, layout=layout)
ply.iplot(fig, filename = "Time Series with Rangeslider")
```
